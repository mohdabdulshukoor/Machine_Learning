```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# TASK 1/2/3

```python
df=pd.read_csv("communities (1).data",header= None)
df
```

```python
#removing..........
X=df.drop(df.columns[:5],axis = 1)
```

```python
# Removes first 5 columns
X
```

```python
df=X.replace("?",np.nan,inplace=True)
df
```

```python
df=X.dropna(axis=0)
```

```python
df
```

```python
X=df.drop(127,axis = 1)
y=df[127]
```

# Polynomial Conversion

```python
# basically polynomial conversion used for removes biasness...
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polynomial_conversion = PolynomialFeatures(degree=3,
                                           include_bias=False)
```

```python
poly_features = polynomial_conversion.fit_transform(X)
poly_features
```

# Train and Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(poly_features,
                                                    y,
                                                    test_size=0.30,
                                                    random_state=101)
```

```python
X_train
```

```python
y_train
```

# Scaling the Data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_train[0]
```

```python
poly_features[0]
```

# Ridge Regression

```python
from sklearn.linear_model import Ridge
```

```python
ridge_model= Ridge(alpha = 1)
```

```python
ridge_model.fit(X_train,y_train)
```

```python
test_predictions = ridge_model.predict(X_test)
test_predictions
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
MAE = mean_absolute_error(y_test, test_predictions)
MSE = mean_squared_error(y_test, test_predictions)
RMSE = np.sqrt(MSE)
```

```python
MAE
```

```python
MSE
```

```python
RMSE
```

```python
plt.plot(X_test,test_predictions,"o",color="orange")
```
