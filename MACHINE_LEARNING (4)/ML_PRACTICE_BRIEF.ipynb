```python
# df["Dates"] dummies...

d = pd.to_datetime(df["Dates"])
df["Year"] = d.dt.year
df["Month"] = d.dt.month
df["DayOfWeek"] = d.dt.day
df.drop(columns=["Dates"],inplace=True)
```

```python
dff=pd.get_dummies(df["Resolution"])
dff
```

## Train Model with Standard Hyperparameters

```python
from sklearn.model_selection import train_test_split
```

```python
X = df1.drop("Category", axis=1)
y = df1["Category"]
```

```python
print(y.value_counts())
```

```python
# split data
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size=0.10,
                                                    random_state=0) 
```

```python
print(y_test.value_counts())
```

```python
from sklearn import tree
```

```python
dtc = tree.DecisionTreeClassifier().fit(X_train, y_train)
```

```python
print(dtc.score(X_test, y_test))
```
