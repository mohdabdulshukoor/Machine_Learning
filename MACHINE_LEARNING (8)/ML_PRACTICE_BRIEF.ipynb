# TASK 3

```python
df=pd.read_csv("communities.data",header= None)
df
```

```python
#removing..........
X=df.drop(df.columns[:5],axis = 1)
```

```python
# Removes first 5 columns
X
```

```python
df=X.replace("?",np.nan,inplace=True)
df
```

```python
df=X.dropna(axis=0)
```

```python
df
```

```python
X=df.drop(127,axis = 1)
y=df[127]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.30,random_state=101)
```

```python
X_train
```

```python
y_train
```

# Creating Model....

```python
from sklearn.linear_model import LinearRegression
```

# Fit / Train the model on the training dataset

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
fig, ax = plt.subplots(figsize=(14,8))
plt.plot(model.coef_)
```

# TASK 4

```python
test_predictions = model.predict(X_test)
test_predictions
```

# Errors

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
MAE = mean_absolute_error(y_test, test_predictions)
MSE = mean_squared_error(y_test, test_predictions)
RMSE = np.sqrt(MSE)
```

```python
#Mean Absolute Error
MAE
```

```python
#Mean Squared Error
MSE
```

```python
#Root Mean Squared Error
RMSE
```

```python
from sklearn import metrics
```

```python
metrics.r2_score(test_predictions,y_test)
```
